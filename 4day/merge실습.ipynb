```python
df1 = pd.DataFrame({"key": list("bbacaab"),  
                    "data1": range(7)})
df2 = pd.DataFrame({"key": list("abd"), 
                    "data2": range(3)}) 
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='key', how='inner') 
```

```python
pd.merge(df1, df2, on='key', how='outer') 
```

```python
pd.merge(df1, df2, on='key', how='left') 
```

```python
pd.merge(df1, df2, on='key', how='right') 
```
