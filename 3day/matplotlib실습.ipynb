```python
sr = pd.Series(np.random.randn(10).cumsum(),
              index=np.arange(0,100,10)) 
sr
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(axis=0),
                  columns=["A", "B", "C", "D"],
                  index=np.arange(0, 100, 10))
df
```

```python
df.plot()
plt.savefig('f2.png')
```
